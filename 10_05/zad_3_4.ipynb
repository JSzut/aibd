## Zadanie 3
Zdefiniowanie zadanej funkcji i sporządzenie wykresów dla argumentów z danego przedziału.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return x**2+5
```

x>-1 && x<1:

```python
x = np.linspace(-1, 1)
plt.plot(x, f(x))
plt.ylabel('y')
plt.xlabel('x')
plt.title("Plot of f(x) from -1 to 1")
plt.legend(['f(x)'])
plt.show()
```

x>-6 && x<6:

```python
x = np.linspace(-6, 6)
plt.plot(x, f(x))
plt.ylabel('y')
plt.xlabel('x')
plt.title("Plot of f(x) from -6 to 6")
plt.legend(['f(x)'])
plt.show()
```

x>0 && x<5:

```python
x = np.linspace(0, 5)
plt.plot(x, f(x))
plt.ylabel('y')
plt.xlabel('x')
plt.title("Plot of f(x) from 0 to 5")
plt.legend(['f(x)'])
plt.show()
```
